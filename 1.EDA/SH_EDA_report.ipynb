```python
import os
import sys

import numpy as np
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity

path = '/root/dev/mlops/0.Data/'

data = pd.read_csv(path + 'beer.csv')
```

```python
data.head()
```

```python
data_pivot = data.pivot_table(index = "아이디", columns = "맥주", values = "평점").fillna(0)
similarity = cosine_similarity(data_pivot.T)
beer_sim = pd.DataFrame(similarity, index = data_pivot.columns, columns = data_pivot.columns)
```

```python
beer_sim.head()
```

```python
def inference(beer_name:str, beer_sim:pd.DataFrame):
    return beer_sim[beer_name].drop(beer_name).sort_values(ascending = False).head(5).index.to_list()
```

```python
inference('Tiger Radler Lemon', beer_sim)
```

```python
def multi_inference(beer_name_list, beer_sim:pd.DataFrame):
    if len(beer_name_list) == 1:
        return inference(beer_name_list, beer_sim)
    avg_sim = beer_sim[beer_name_list].mean(axis = 1)
    recommendations = avg_sim.drop(beer_name_list).sort_values(ascending = False).head(5).index.to_list()
    return recommendations
```

```python
multi_inference(["FiLite", "Hite Prime Max"], beer_sim)
```
